```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.cluster import KMeans
from lifetimes import BetaGeoFitter , GammaGammaFitter
from lifetimes.utils import summary_data_from_transaction_data
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
```

```python
def remove_outliers(a):
    amin=a.quantile(0.25)
    amax=a.quantile(.75)
    iqr=amax-amin
    return a[(a>(amin-(1.5*iqr)))&(a<(amax+(1.5*iqr)))]
```

```python
df.dropna(subset=["CustomerID"],inplace=True)
```

```python
df=df[(df["Quantity"]>=0)&(df["UnitPrice"]>=0)]
```

```python
df["TotalPriceStock"]=df["Quantity"]*df["UnitPrice"]
```

```python
df1=df.groupby(["CustomerID","InvoiceNo"]).agg(Date=("InvoiceDate","first"),TotalPriceStock=("TotalPriceStock","sum"))
    
```

```python
customer_df=df1.reset_index(level="InvoiceNo").groupby("CustomerID").agg(AverageInvoicePrice=("TotalPriceStock","mean"),
Monetary=("TotalPriceStock","sum"),Frequency=("TotalPriceStock","count"),
First=("Date","min"),
Last=("Date","max"))
snapshot_date=df["InvoiceDate"].max()+pd.to_timedelta(1,unit="D")
startdate=df["InvoiceDate"].min()-pd.to_timedelta(1,unit="D")
customer_df["Recency"]=customer_df["Last"].apply(lambda x:(snapshot_date-x).days)
customer_df["Arrival"]=customer_df["First"].apply(lambda x:(x-startdate).days)

original_cust_df=customer_df.copy()
```

```python
freq_cap=original_lifetimes_df["frequency"].quantile(0.99)
lifetimes_df = lifetimes_df[lifetimes_df['frequency'] <= freq_cap]
```

```python
print(lifetimes_df[lifetimes_df['recency'] > lifetimes_df['T']])
print(lifetimes_df[(lifetimes_df['frequency'] == 0) & (lifetimes_df['recency'] > 0)])
print(lifetimes_df[lifetimes_df['T'] == 0])
```

```python
bgf = BetaGeoFitter(penalizer_coef=0.05) # Penalizer helps with convergence

gg_data = lifetimes_df[lifetimes_df['frequency'] > 0]
correlation = gg_data[['frequency', 'recency']].corr().iloc[0, 1] 
print(correlation)
ggf = GammaGammaFitter(penalizer_coef=10)
ggf.fit(gg_data['frequency'], gg_data['monetary_value'])

bgf.fit(lifetimes_df['frequency'], lifetimes_df['recency'], lifetimes_df['T'])
```

```python
print(gg_data[['frequency', 'monetary_value']].corr().iloc[0, 1] )
```

```python
N_DAYS = 90
print(f"Calculating CLV for the next {N_DAYS} days...")
    
clv_predictions = ggf.customer_lifetime_value(
    bgf,
    lifetimes_df['frequency'],
    lifetimes_df['recency'],
    lifetimes_df['T'],
    lifetimes_df['monetary_value'],
    time=N_DAYS / 30,  # 'time' is in months
    discount_rate=0.01 # Monthly discount rate (e.g., 1%)
)
    
clv_df = clv_predictions.to_frame(name='CLV')
clv_df = clv_df.reset_index()
    
print("CLV prediction complete.")
print(clv_df.head())
    
print(clv_df)
```

```python
from lifetimes import BetaGeoFitter

# You can then create an instance of the model
bgnbd_model = bgf
clv_summary_df=lifetimes_df
print(bgnbd_model)
```

```python
t_days = 90
clv_summary_df['predicted_purchases_90d'] = bgnbd_model.conditional_expected_number_of_purchases_up_to_time(
    t=t_days,
    frequency=clv_summary_df['frequency'],
    recency=clv_summary_df['recency'],
    T=clv_summary_df['T']
)
        
        # Get the monetary proxy
        # For f=0 customers (one-time buyers), use the population median
median_monetary_value = clv_summary_df['monetary_value'].median()
        
        # For f>0 customers (repeat buyers), use their own historical average
clv_summary_df['monetary_proxy'] = np.where(
        clv_summary_df['frequency'] > 0,
        clv_summary_df['monetary_value'],
        median_monetary_value
)
        
        # Calculate simple CLV = E[Purchases] * E[Monetary Value]
        # (We ignore discounting in this simpler, more stable heuristic)
clv_prediction = clv_summary_df['predicted_purchases_90d'] * clv_summary_df['monetary_proxy']

    # 5. Final safety net: Clip any tiny negative numerical artifacts to 0
clv_prediction = clv_prediction.clip(lower=0)
print(clv_prediction)
```

```python
min_samples=6
nn = NearestNeighbors(n_neighbors=min_samples)
nn.fit(customer_df[["Frequency","Recency","Monetary"]])
distances, indices = nn.kneighbors(customer_df[["Frequency","Recency","Monetary"]])
k_distances = np.sort(distances[:, min_samples - 1], axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_distances)
plt.title(f'K-Distance Graph (k={min_samples})')
plt.xlabel('Data Points (sorted by distance)')
plt.ylabel(f'{min_samples}-th Nearest Neighbor Distance (eps)')
plt.grid(True)
plt.show()
```

```python


# 1. Our chosen parameters from the analysis
optimal_eps = 0.1
optimal_min_samples = 6

# 2. Run DBSCAN
dbscan = DBSCAN(eps=optimal_eps, min_samples=optimal_min_samples)
clusters = dbscan.fit_predict(customer_df[["Frequency","Recency","Monetary"]])

# 3. Add the cluster labels back to your original DataFrame
original_cust_df['Cluster_labels'] = clusters

# 4. Check the results
print(original_cust_df['Cluster_labels'].value_counts())
```
